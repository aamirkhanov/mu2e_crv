```python
def default_flatname(branchname, fieldname, index):
    """This replaces the dot in the column names with underscores"""
    out = branchname.replace(".", "_")
    if not isinstance(branchname, str):
        out = branchname.decode("utf-8")
    if fieldname is not None:
        fieldname.replace(".", "_")
        out += "_" + fieldname
    if index != ():
        out += "[" + "][".join(str(x) for x in index) + "]"
    return out
```

```python
vars_to_save = ['evtinfo__eventid',
        'evtinfo__runid',
        'evtinfo__subrunid',
        'evtinfo__nprotons',
        'de_status',
        'de_pdg',
        'de_nhits',
        'de_t0',
        'de_t0err',
        'deent_mom',
        'deent_momerr',
        'deent_d0',
        'deent_p0',
        'deent_om',
        'deent_z0',
        'deent_td',
        'dequal_nquals',
        'dequal_TrkPID',
        'dequal_TrkPIDCalib',
        'dequal_TrkQual',
        'dequal_TrkQualCalib',
        'demc_pdg',
        'trigbits',
        'ue_status',
        'ue_pdg',
        'ue_nhits',
        'ue_t0',
        'ue_t0err',
        '_totalPEs',
        '_nHitCounters',
        '_totalEnergyDeposited',
        '_minPathLayer',
        '_maxPathLayer',
        '_nHitCounters',
        '_x',
        '_y',
        '_z',
        '_crvSectorNumber',
        '_crvSectorType',
        'is_high',
        'lyield',
        'crvinfo__dT',
        'deent_d0_om',
        'is_triggered',
        'crvinfo__crvSectorType',
        'crvinfo__x',
        'crvinfo__y',
        'crvinfo__z',
        'crvinfo__timeWindowStart',
        'crvinfo__timeWindowEnd',
        'crvinfo__PEs',
        'crvinfo__nCoincidenceHits',
        'crvinfomc__x',
        'crvinfomc__y',
        'crvinfomc__z',
        'crvinfomc__time',
        'crvinfomc__depositedEnergy',
        'bestcrv']        

def make_df(trkana, tag_high, lyield):
  df_eve = trkana.pandas.df(['evtinfo.', 'de', 'deent', 'dequal', 'demc', 'trigbits', 'ue', 'crvsummary*', '_*'], flatname=default_flatname, flatten=True)
  #df_eve = trkana.pandas.df(eve_vars, flatname=default_flatname, flatten=True)
  #df_crv = trkana.pandas.df(['crvinfo.*', 'crvinfomc.*', 'bestcrv'], flatname=default_flatname, flatten=True)
  df_crv = trkana.pandas.df(['crvinfo.*', 'bestcrv'], flatname=default_flatname, flatten=True)
  #df_crv = trkana.pandas.df(crv_vars, flatname=default_flatname, flatten=True)
  df_crv = df_crv[df_crv.index.get_level_values('subentry') == df_crv["bestcrv"]]

  pd.merge(df_eve, df_crv, how="left", on='entry')
  df_ = df_eve.merge(df_crv, how="left", on="entry")
  df_['is_high'] = tag_high
  df_['lyield'] = int(lyield)
  df_["crvinfo__dT"] = df_["crvinfo__timeWindowStart"] - df_["de_t0"]
  df_['deent_d0_om'] = df_["deent_d0"]+2.0/df_["deent_om"]
  df_["is_triggered"] = (df_["trigbits"].values & 0x208) > 0
  
  print("before deleting " + str(df_.shape))
  
  for col in df_.columns:
    if col not in vars_to_save:
      del df_[col]

  print("after deleting " + str(df_.shape))
  
  df_ = df_.fillna(-999999)
#  df = pd.concat([df, df_], axis=0)
  return df_
```

```python
#ratio_vals = ratio_vals_no_norm
min_eff_constr = 1e-7

def get_efficiency(_ratio_vals):

  ratio_vals = _ratio_vals


  eff = {}
  for lyield in ratio_vals:
    eff[lyield] = ratio_vals[lyield][1][0]/ratio_vals[lyield][2][0]

  for lyield in ratio_vals:
    eff[lyield] = 1-ratio_vals[lyield][1][0]/ratio_vals[lyield][2][0]
    for index, sec_eff in enumerate(eff[lyield]):
      if sec_eff == 0 or sec_eff < min_eff_constr:
        eff[lyield][index] = min_eff_constr

  _eff_by_sec = np.empty((22, 6), float)
  #for lyield in ratio_vals:
  #  eff[lyield] = 1-ratio_vals[lyield][1][0]/ratio_vals[lyield][2][0]

  for sec in range(0,len(sectorNames)):
    for index, lyield in enumerate(eff):
      _eff_by_sec[sec][index] = eff[lyield][sec]

  return _eff_by_sec

```
